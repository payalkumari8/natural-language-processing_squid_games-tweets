```python
import spacy
import pandas as pd
```

```python
import opendatasets as od
download_url = 'https://www.kaggle.com/deepcontractor/squid-game-netflix-twitter-data/code'

od.download(download_url)
```

```python
data_filename = './squid-game-netflix-twitter-data/tweets.csv'
```

```python
#read the tweets 
tweets = pd.read_csv(data_filename)
tweets
```

```python
tweets.head()
```

```python
tweets.describe()
```

```python
#Load the pre-trained model
import en_core_web_sm
#nlp = en_core_web_sm.load()
nlp = spacy.load("en_core_web_sm")
```

```python
#Extract text from Pandas Colum

text=str(tweets['text'])
doc = nlp(text)
```

```python
#Parts of Speech


for token in doc:
    print(token.text, token.pos_)
    
```

```python
#Named Entity Recognition


spacy.displacy.render(doc,style='ent',jupyter=True)
```

```python
#Extracting a particular entity type


for ent in doc.ents:
    if ent.label_ == 'PERSON':
        print(ent.text,ent.label_)
```
