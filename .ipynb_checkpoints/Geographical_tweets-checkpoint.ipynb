# Squid Games Geographical Distribution of Tweets 

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
import opendatasets as od
download_url = 'https://www.kaggle.com/deepcontractor/squid-game-netflix-twitter-data/code'

od.download(download_url)
```

```python
data_filename = './squid-game-netflix-twitter-data/tweets.csv'
```

```python
data = pd.read_csv(data_filename)
data
```

```python
data.head(10)
```

```python
data.isnull().count()
```

```python
data = data.dropna()
print('Total Number of Unique Locations Tweeting :',len(data['user_location'].unique()))
```

```python
#Top 5 Locations Tweeting Squid Games
top5 = data['user_location'].value_counts()
top5 = top5[0:5]
top5 = top5.to_frame().reset_index()
top5.rename(columns = {'index':'location','user_location':'counts'}, inplace = True) 





```

```python
sns.barplot(y="location", x="counts", data=top5)
sns.set(rc={'figure.figsize':(8,5)})

```

```python
#Top 15 Locations Tweeting Squid Games
top15 = data['user_location'].value_counts()
top15 = top15[0:15]
top15 = top15.to_frame().reset_index()
top15.rename(columns = {'index':'location','user_location':'counts'}, inplace = True) 




```

```python
sns.barplot(y="location", x="counts", data=top15)
sns.set(rc={'figure.figsize':(15,10)})
```
